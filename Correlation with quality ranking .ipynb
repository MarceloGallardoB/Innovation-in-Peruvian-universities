```python
import pandas as pd

# Leer el archivo Excel
df_excel = pd.read_excel('top sunedu universidades.xlsx')
df_excel
```

```python
df_excel = df_excel.dropna(subset=['University', 'Ranking', 'Indicator'])

# Calcular la correlación entre las dos últimas columnas
correlation = df_excel['Ranking'].corr(df_excel['Indicator'])

# Imprimir la correlación
print("Correlación entre las dos últimas columnas:", correlation)
```

```python
correlation = 0.8057823569018411  # Reemplaza este valor con el resultado real de la correlación

# Formatear la correlación en formato LaTeX
latex_correlation = f"Correlación entre las dos últimas columnas: {correlation:.4f}"

# Imprimir el resultado en formato LaTeX
print(latex_correlation)
```

```python
latex_table = df_excel.to_latex(index=False, caption="DataFrame in LaTeX Format")

# Print the LaTeX table
print(latex_table)
```
